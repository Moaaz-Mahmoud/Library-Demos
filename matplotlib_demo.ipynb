## Importing Libraries

## _y = f(x)_ Example

```python
plt.style.use('seaborn')
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.xlim(-2*np.pi, 2*np.pi)
plt.plot(x, 1/np.cosh(x**3)+np.tanh(x)+0.1*x+0.2*np.sin(x))
plt.title('1/np.cosh(x**3)+np.tanh(x)+0.1*x+0.2*np.sin(x)')
```

## Parametric Examples
### Example 1

```python
#General settings
t = np.linspace(0.2, 2*np.pi-0.2, 100)
plt.axis('equal')
plt.title('Capped Circle')
plt.xlabel('x')
plt.ylabel('y')
color=(0.4, 0.4, 0.8)

#The open circle
x = np.sin(t)
y = np.cos(t)
plt.axis([-2, 2, -2, 2])
plt.plot(x, y, color=color)

#The cap
t = np.linspace(0.8, 1.2, 2)
plt.plot([0.2]*len(t), t, color=color)
plt.plot([-0.2]*len(t), t, color=color)
plt.plot([-0.2, 0.2], [1.2]*2, color=color)
```

### Example 2

```python
plt.title('Egg (Parametric)')
plt.axis('equal')
plt.style.use('seaborn')
plt.axis([-1, 3, -1, 3])

t = np.linspace(-np.pi, np.pi, 400)
a = 0.83
b = 1
x = 0.9 * a * np.sin(t + np.sin(t)/4) + b
y = 1 - 1.2 * a * np.cos(t)

plt.plot(x, y, '-c')
```

## Random Scatter Example

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
cmap='viridis')
plt.colorbar(); # show color scale
```

## Three-Dimensional Examples
### Example 1

```python
figure = plt.figure()
ax = plt.axes(projection='3d')
plt.title('3D Spring')
z = np.linspace(0, 40, 800)
y = np.sin(z)
x = np.cos(z)
ax.plot3D(x, y, z)
plt.show()
```

### Example 2

```python
figure = plt.figure()
ax = plt.axes(projection='3d')
y = np.linspace(0, 40, 500)
z = np.sin(y)
x = np.cos(y) * z
ax.plot3D(x, y, z)
plt.show()
```

## Legend Example

```python
fig, ax = plt.subplots()

t = np.linspace(0, 4*np.pi, 500)
x = t - 2*np.sin(t)
y = np.cos(2*t) + t

ax.plot(x, y, label='crv1', color=(0.8, 0.2, 0.5))
ax.plot(y, x, label='crv2', color=(0.5, 0.2, 0.8))

ax.legend(frameon=True, fancybox=True, loc='upper left',
 framealpha=1, shadow=True, borderpad=0.5)
```

## Gradient Color

```python
for i in range(10):
    x = np.linspace(i*np.pi/2, (i+1)*np.pi/2, 200)
    y = np.sin(x)
    plt.plot(x, y, color=(i/100, 1-i/10, i/10))
```
