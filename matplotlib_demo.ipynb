## Parametric Examples
### Example 1

```python
#General settings
t = np.linspace(0.2, 2*np.pi-0.2, 100)
plt.axis('equal')
plt.title('Capped Circle')
plt.xlabel('x')
plt.ylabel('y')
color=(0.4, 0.4, 0.8)

#The open circle
x = np.sin(t)
y = np.cos(t)
plt.axis([-2, 2, -2, 2])
plt.plot(x, y, color=color)

#The cap
t = np.linspace(0.8, 1.2, 2)
plt.plot([0.2]*len(t), t, color=color)
plt.plot([-0.2]*len(t), t, color=color)
plt.plot([-0.2, 0.2], [1.2]*2, color=color)
```

### Example 2

```python
plt.title('Egg (Parametric)')
plt.axis('equal')
plt.style.use('seaborn')
plt.axis([0, 4, -0.5, 3])

t = np.linspace(-np.pi, np.pi, 400)
a = 0.83
b = 1

x = 0.9 * a * np.sin(t + np.sin(t)/4) + b
y = 1 - 1.2 * a * np.cos(t)

plt.plot(x, y, '-c')
plt.plot(x+1.7, y+0.4, '--m') #Translated
```

## Legend Example

```python
fig, ax = plt.subplots()

t = np.linspace(0, 4*np.pi, 500)
x = t - 2*np.sin(t)
y = np.cos(2*t) + t

ax.plot(x, y, label='crv1', color=(0.8, 0.2, 0.5))
ax.plot(y, x, label='crv2', color=(0.5, 0.2, 0.8))

ax.legend(frameon=True, fancybox=True, loc='upper left',
 framealpha=1, shadow=True, borderpad=0.5)
```

## Multiple Subplots Example

```python
sech = lambda x: 1/np.cosh(x)

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4])
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4])

t = np.linspace(0, 2*np.pi, 720)
x = np.tanh(t)
y = sech(t) + np.sin(t)

ax1.plot(x, y)
ax2.plot(y, x)
```

### A Wireframe Example

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

a = 40
x = np.linspace(-10, 10, a)
y = np.linspace(-10, 10, a)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)

ax.plot_wireframe(X, Y, Z)
ax.view_init(60, 45)
```

### A Surface Example

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

a = 40
x = np.linspace(-10, 10, a)
y = np.linspace(-10, 10, a)
X, Y = np.meshgrid(x, y)
Z = np.sin(0.6*X) + np.cos(0.6*Y)

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.view_init(45, 30)
```

### Cylindrical-Coordinates Examples

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)
X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
cmap='viridis', edgecolor='none')
```

```python
r = np.linspace(0, 5, 25)
theta = np.linspace(0, 2*np.pi, 27)
r, theta = np.meshgrid(r, theta)

X = r * np.cos(theta)
Y = r * np.sin(theta)
Z = np.sin(X) - np.cos(Y) + np.sqrt(X**2 + Y**2)

ax = plt.axes(projection='3d')
ax.set_zlim3d(-2, 8)
ax.plot_surface(X, Y, Z, cmap='viridis')
```
