## Importing Libraries

```python
import re
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

## Code

```python
plt.style.use('seaborn')
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
plt.xlim(-2*np.pi, 2*np.pi)
plt.plot(x, 1/np.cosh(x**3)+np.tanh(x)+0.1*x+0.2*np.sin(x))
plt.title('1/np.cosh(x**3)+np.tanh(x)+0.1*x+0.2*np.sin(x)')
```

```python
t = np.linspace(0.2, 2*np.pi-0.2, 100)
plt.axis('equal')
plt.title('Capped Circle')
plt.xlabel('x')
plt.ylabel('y')
color=(0.3, 0.4, 0.8)

x = np.sin(t)
y = np.cos(t)
plt.axis([-2, 2, -2, 2])
plt.plot(x, y, color=color)

t = np.linspace(0.8, 1.2, 2)
plt.plot([0.2]*len(t), t, color=color)
plt.plot([-0.2]*len(t), t, color=color)
plt.plot([-0.2, 0.2], [1.2]*2, color=color)
```

```python
plt.title('Egg (Parametric)')
plt.axis('equal')
plt.style.use('seaborn')
plt.axis([-1, 3, -1, 3])

t = np.linspace(-np.pi, np.pi, 40)
a = 0.83
b = 1
x = 0.9 * a * np.sin(t + np.sin(t)/4) + b
y = 1 - 1.2 * a * np.cos(t)

plt.plot(x, y, '-c')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
cmap='viridis')
plt.colorbar(); # show color scale
```

```python
figure = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 30, 700)
y = np.sin(z)
x = np.cos(z)
ax.plot3D(x, y, z)
plt.show()
```

```python
figure = plt.figure()
ax = plt.axes(projection='3d')
y = np.linspace(0, 40, 500)
z = np.sin(y)
x = np.cos(y) * z
ax.plot3D(x, y, z)
plt.show()
```
